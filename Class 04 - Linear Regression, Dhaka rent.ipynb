```python
# We are now dividing two columns. We will put the whole area column in x and the whole rent column in y.


x = df[['area']]             # x cannot be one dimentional. To make it two dimentional we have to put double bracket.
                             # x = df.drop('rent', axis=1),if we drop rent column, only area column remains
                             # area column kept inside x variable
```

```python
y = df[['rent']]             # rent column kept inside y variable
                             # y = df.drop('area', axis =1 )
```

```python
x.head()                     # To show first 5 x values = area column
```

```python
y.head()                      # To show first 5 y values = rent column 
```

```python
# !pip install sklearn
```

```python
#Now after dividing the columns, we will apply linear regression

from sklearn.linear_model import LinearRegression      # LinearRegression will be imported from linear_model and 
                                                    # linear_model, a module inside sklearn will be imported from sklearn
```

```python
reg = LinearRegression()                                # Creating object for LinearRegression
```

```python
reg.fit(x,y)                                         # this line is used to calculate everything like x bar, m and c
                                                     # training the model
```

```python
reg.coef_                             # To calculate the coefficient,m
```

```python
reg.intercept_                         # To calculate the y-intercept,c
```

# Ignore Warning

```python
import warnings as wr
wr.filterwarnings('ignore')
```

```python
reg.predict([[4000]])           # Predicted rent(y) for x = 4000
```

```python
reg.predict([[3000]])         # Predicted rent(y) for x = 3000
```

```python
residuals = 52408.2489744 - 52100       # Loss/error/residuals = |Actual rent- predicted rent|
```

```python
#y = m*x+c

result = ((reg.coef_)*3000) + reg.intercept_
```

```python
result                        # Matched with line 34
```

```python
reg.predict(x)
```

```python
# All the x values will have a predicted rent and by using df['predicted_rent']
# We are creating a new column for predicted rent.

df['predicted_rent'] = reg.predict(x)
```
